```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ukuran = ['50', '100', '150', '200', '250', '300', '400', '500']
harga = [700, 1400, 2100, 2800, 3500, 4200, 4900, 5600]
ax.bar(ukuran,harga)
plt.grid()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
ukuran = ['50', '100', '150', '200', '250', '300', '400', '500']
harga = [700, 1400, 2100, 2800, 3500, 4200, 4900, 5600]
ax.pie(ukuran,labels=harga,autopct='%1.2f%%')
plt.title('Piechart')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ukuran = [50, 100, 150, 200, 250, 300, 400, 500]
harga = [700, 1400, 2100, 2800, 3500, 4200, 4900, 5600]

ax.scatter(ukuran,harga,color='blue')

ax.set_xlabel('ukuran')
ax.set_ylabel('harga')
ax.set_title('Scatter Plot')

plt.grid()
plt.show()
```
